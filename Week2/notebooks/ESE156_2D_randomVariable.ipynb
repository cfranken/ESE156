The multi-dimensional Gaussian distribution of a random variable $x$ with mean $\mu$ is 
$$f(x_1,\ldots,x_k) = \frac{\exp\left(-\frac{1}{2}(x-\mu)^TS^{-1}(x-\mu)\right)}{\sqrt{(2\pi)^k \det{S}}}$$

```python
x1 = np.arange(-10,10,0.1)
x2 = np.arange(-10,10,0.1)

Sa = np.array(np.mat('1 1.4; 1.4 2'))
Sa2 = np.array(np.mat('1 0.5; 0.5 2'))

N = len(x1)
invSa = np.linalg.inv(Sa)
invSa2 = np.linalg.inv(Sa2)

f = np.zeros((N,N))
f2 = np.zeros((N,N))

for i in range(N):
    for j in range(N):
        x = np.asarray([x1[i],x2[j]])
        
        f[i,j] = np.exp(-0.5*x.T.dot(invSa).dot(x))/np.sqrt((2*np.pi)**2.0*np.linalg.det(Sa))
        f2[i,j] = np.exp(-0.5*x.T.dot(invSa2).dot(x))/np.sqrt((2*np.pi)**2.0*np.linalg.det(Sa2))
```

```python
print(invSa)
print(invSa2)
```

```python
figsize(10,10)
plt.contour(x1,x2,f)
plt.contour(x1,x2,f2)
plt.colorbar()
plt.xlim((-5,5))
plt.ylim((-5,5))
```
